```python
import pandas as pd
import pyodbc
from scipy import stats

# 1. DATABASE CONNECTION CONFIGURATION
server = r"(localdb)\MSSQLLocalDB"
database = "LaptopPriceAnalysis"
driver = "ODBC Driver 17 for SQL Server"

connection_string = (
    f"DRIVER={{{driver}}};"
    f"SERVER={server};"
    f"DATABASE={database};"
    "Trusted_Connection=yes;"
)

def get_and_clean_data():
    """
    Fetches raw data from SQL Server, applies standardization mapping, 
    and saves the cleaned dataframe as a pickle file.
    """
    try:
        # Establish database connection
        conn = pyodbc.connect(connection_string)
        print("Connection Successful")

        # Execute query and load data into a DataFrame
        query = "SELECT * FROM dbo.Laptops_Final_Analysis"
        df = pd.read_sql(query, conn)
        
        # 2. CATEGORICAL DATA STANDARDIZATION (Mapping)
        brand_mapping = {
            "LENOVO": "Lenovo", "ACER": "Acer", "DELL": "Dell", "ASUS": "Asus"
        }
        
        ssd_mapping = {
            "16GB-512GB SSD": "512GB", "8GB-512GB SSD": "512GB", "512GB SSD": "512GB",
            "32GB-512GB SSD": "512GB", "512GBSSD": "512GB", "16GB -512GB SSD": "512GB",
            "12GB-512GB SSD": "512GB", "8GB-512GBSSD": "512GB",
            "8GB-256GB SSD": "256GB", "256GB SSD": "256GB", "4GB - 256GB SSD": "256GB"
        }
        
        # Apply replacements
        df["Brand"] = df["Brand"].replace(brand_mapping)
        df["SSD"] = df["SSD"].replace(ssd_mapping)
        df["Core"] = df["Core"].replace("Core i513420", "Core i5")
        
        # 3. SAVE AS MASTER DATA FILE (Pickle format for persistence)
        df.to_pickle("master_laptop_data.pkl")
        
        print(f"Success: {len(df)} rows cleaned and processed.")
        print("Master file saved as 'master_laptop_data.pkl'")
        
        conn.close()
        return df

    except Exception as e:
        print(f"An error occurred: {e}")

# Run data preparation script
master_df = get_and_clean_data()
```
