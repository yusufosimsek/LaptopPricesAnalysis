```python
import pandas as pd

# Load the master dataset from pickle format
df = pd.read_pickle("master_laptop_data.pkl")

# Convert categorical columns to 'category' dtype for memory efficiency
df["Brand"]     = df["Brand"].astype("category")
df["Core"]      = df["Core"].astype("category")
df["SSD"]       = df["SSD"].astype("category")
df["Series"]    = df["Series"].astype("category")
df["Processor"] = df["Processor"].astype("category")
df["RAM"]       = df["RAM"].astype("category")
df["Storage"]   = df["Storage"].astype("category")

# Ensure numeric columns are properly typed; invalid values become NaN
df["Rating"]           = pd.to_numeric(df["Rating"],           errors="coerce")
df["Reviews"]          = pd.to_numeric(df["Reviews"],          errors="coerce")
df["Actual_Price"]     = pd.to_numeric(df["Actual_Price"],     errors="coerce")
df["Discounted_Price"] = pd.to_numeric(df["Discounted_Price"], errors="coerce")

# Drop the surrogate key — not needed for dashboard visualizations
df = df.drop(columns=["Laptop_ID"])

# Export to CSV with BOM-prefixed UTF-8 encoding for Power BI compatibility
df.to_csv("laptop_dashboard.csv", index=False, encoding="utf-8-sig")

print(df.dtypes)
print(f"\n{len(df)} rows exported successfully.")
```
